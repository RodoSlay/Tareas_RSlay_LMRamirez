Sustituyendo$a=40,b=49,\mu=44,\sigma^2=9$, obtenemos las siguientes funciones:
$$\frac{40\beta+49\alpha}{\alpha+\beta}-44=0$$
$$\frac{\alpha\beta(49-40)^2}{(\alpha+\beta)^2(1+\alpha+\beta)}-9=0$$

### Ejercicio 1: (enunciado)

```python
# Importamos librerías a trabajar en todas las simulaciones
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as st     # Librería estadística
from scipy.optimize import fsolve
%matplotlib inline
```

```python
#RSR
def f1(p):
    alfa,beta=p
    return ((40*beta+49*alfa)/(alfa+beta)-44, (alfa*beta*(49-40)**2)/((alfa+beta)**2*(1+alfa+beta))-9)

alfa,beta = fsolve(f1, (44,9))
alfa,beta
```

```python
#LMRR
def f1(p):
    alfa,beta=p
    return ((40*beta+49*alfa)/(alfa+beta)-44, (alfa*beta*(49-40)**2)/((alfa+beta)**2*(1+alfa+beta))-9)

alfa,beta = fsolve(f1, (44,9))
alfa,beta
```

Obtenemos los valores de $\alpha$ y de $\beta$, para comprobar se van a evaluar las funciones y comprobar que en realidad tienden a cero.

```python
(40*beta+49*alfa)/(alfa+beta)-44, (alfa*beta*(49-40)**2)/((alfa+beta)**2*(1+alfa+beta))-9
```

Observamos que efectivamente tienden a cero.

```python
#RSR
a=alfa
b=beta
x = np.linspace(40,49,100)
r = st.beta.pdf(x,a,b,loc=40,scale=9)
r_acum = st.beta.cdf(x,a,b,loc=40,scale=9)
# Graficamos
fig,(ax1,ax2) = plt.subplots(1,2)
fig.set_figwidth(15)
ax1.plot(x,r,'o',markersize=3)
ax1.set_title('Densidad de probabilidad')

# Probabilidad acumulada
ax2.plot(x,r_acum,'o',markersize=3)
ax2.set_title('Distribución acumulada')

plt.show()
```

```python
#LMRR
a=alfa
b=beta
x = np.linspace(40,49,100)
r = st.beta.pdf(x,a,b,loc=40,scale=9)
r_acum = st.beta.cdf(x,a,b,loc=40,scale=9)
# Graficamos
fig,(ax1,ax2) = plt.subplots(1,2)
fig.set_figwidth(15)
ax1.plot(x,r,'xr',markersize=3)
ax1.set_title('Densidad de probabilidad')

# Probabilidad acumulada
ax2.plot(x,r_acum,'xr',markersize=3)
ax2.set_title('Distribución acumulada')

plt.show()
```

Ahora se saca la probabilidad de que $P(T\leq d)$.

```python
#RSR
d = np.linspace(40,49,10)
st.beta.cdf(d,a,b,loc=40,scale=9)
```

```python
#LMRR
d = np.linspace(40,49,10)
st.beta.cdf(d,a,b,loc=40,scale=9)
```

Aquí podemos ver las probabilidades para diferentes valores de d = 40,41,42,43,44,45,46,47,48 y 49.
